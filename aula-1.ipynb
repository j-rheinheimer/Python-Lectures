## Um pouco de teoria

A linguagem de Programação Python foi concebida em 1989 por Guido van Rossum no Instituto de Pesquisa Nacional para Matemática e Ciência da Computação (CWI). O Python é uma linguagem moderna, de alto nível, interpretada, de tipagem dinâmica e orientada a objetos (Confuso, né? Com o tempo vamos entender cada um desses termos).  
Python é considerada uma linguagem de alto nível pelo fato da escrita do código ser muito próxima da forma como escrevemos no dia a dia, sem ser a necessário o uso de muito caractéres "estranho" ou estruturas complexas. Um exemplo desse "alto nível" pode ser visto nas células de código acima, onde bastou o comando "print" (Literalmente traduzido como "escreva/mostre") com o texto que queríamos mostrar na tela, definitivamente simples de entender.  
Em outras linguagens (Como por exemplo o C, pai do Python) o nosso "Hello World" teria de ser escrito da seguinte forma:  

```  
#include <stdio.h>  
int main() {  
  printf("Hello, World!");  
  return 0;  
}  
```  

Complexo, não?  
  
O fato da linguagem ser chamada de interpretada esta ligada a forma como o computador irá lidar com o código. Diferentemente de linguagens como C, C++, Fortran e muitas outras, o Python é uma linguagem interpretada (Como Javascript, R, Ruby e Lua). Este fato faz com que no momento da execução do código não haja a necessidade da compilação.  
O que acontece no momento da execução de um código de Python é a interpretação do código em tempo real para a máquina, como se  o Interpretador Python fosse um tradutor, traduzindo linha por linha de nosso código enquanto a máquina executa os comandos. Em linguagens compiladas primeiro há toda a transcrição do código para linguagem de máquina e após isso a execução.
